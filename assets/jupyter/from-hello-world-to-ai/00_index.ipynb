Welcome. This course takes you from your very first line of Python all the way to fine-tuning a modern Transformer model, the kind of architecture behind today's most capable AI systems. Every notebook runs **directly in your browser**: there is nothing to install and nothing to download.

Each concept is first explained in depth, then implemented **by hand**, and only then handed over to a professional library. That order is deliberate: it is what turns these systems from black boxes into things you genuinely understand.

## How to use these notebooks

- Work through them **in order**, each one builds on the last.
- **Run every code cell yourself** (click it and press `Shift + Enter`). Programming is learned by doing.
- Attempt every exercise **before** opening the solution.
- Some later notebooks install extra packages on first use (you will see a short `piplite` instruction at the top). Heavy libraries such as PyTorch may take a little while in the browser.

## A note on the learning philosophy

There is a single thread running through the whole course, the humble **dot product**. It appears in a NumPy array operation, inside a neuron, across a convolutional filter, and at the heart of attention. Modern AI is built from a surprisingly small set of ideas, combined and scaled. By the end, you will see that clearly.
