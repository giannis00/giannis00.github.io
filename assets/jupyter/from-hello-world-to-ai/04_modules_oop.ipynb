## 1. Modules: standing on the shoulders of others

### Intuition first
You rarely build everything from scratch. Python ships with a large **standard library** of ready-made tools, organised into **modules**. Importing a module is like opening a toolbox and pulling out exactly the tool you need.

### How importing works
The `import` keyword loads a module so you can use its functions. Let us look at `math` and `random`.

````python
import math

print(math.sqrt(16))      # square root -> 4.0
print(math.pi)            # the constant pi
print(math.factorial(5))  # 5! = 120
print(math.ceil(4.1))     # round up -> 5
print(math.floor(4.9))    # round down -> 4
````

````python
import random

random.seed(42)   # fix the seed so results are REPRODUCIBLE (same every run)
print(random.randint(1, 6))            # a dice roll between 1 and 6
print(random.choice(["A", "B", "C"]))   # a random pick from a list

deck = [1, 2, 3, 4, 5]
random.shuffle(deck)                    # shuffle a list in place
print("Shuffled:", deck)
````

### Importing specific names
Instead of importing a whole module, you can import just the parts you need. Then you use them without the module prefix.

````python
from math import sqrt, pi

print(sqrt(25))   # no "math." prefix needed now
print(pi)
````

### Aliasing with `as`
We often import a library under a short **alias**. This is not a quirk: it is the universal convention you will see in every data-science file.

````python
import statistics as stats
print(stats.mean([10, 8, 9, 7]))
print(stats.median([10, 8, 9, 7]))

# In the coming modules you will write, by the same logic:
#   import numpy as np
#   import pandas as pd
````

> 🧠 **A note on reproducibility.** `random.seed(42)` makes the 'random' numbers come out the same every time the code runs. In science and machine learning this is essential: it lets others reproduce your exact results. We will set seeds throughout this course.

## 2. From functions to objects

### The motivation
So far we have kept **data** (variables) and **actions** (functions) separate. But in the real world, data and the actions that operate on it naturally belong together. A bank account *has* a balance **and** *can* deposit or withdraw. A student *has* grades **and** *can* compute their average.

**Object-Oriented Programming (OOP)** bundles related data and behaviour into a single unit called an **object**.

### Class vs object: the key distinction
- A **class** is a *blueprint*: a general description of something.
- An **object** (or *instance*) is a *concrete thing* built from that blueprint.

Analogy: a class is the architectural plan for a house; objects are the actual houses built from that one plan. One blueprint, many houses, each with its own address and colour, but all following the same design.

## 3. Your first class

Two pieces of vocabulary you will meet immediately:
- **`__init__`** is a special method (the *constructor*) that runs automatically whenever a new object is created. It sets up the object's starting data.
- **`self`** refers to *the particular object being worked on*. It is how each object keeps track of its own data. Every method's first parameter is `self`.

````python
class Student:
    def __init__(self, name, year):
        self.name = name      # these are ATTRIBUTES (the object's data)
        self.year = year
        self.grades = []

    def add_grade(self, grade):   # this is a METHOD (an action)
        self.grades.append(grade)

    def average(self):
        if not self.grades:
            return 0
        return sum(self.grades) / len(self.grades)

# Create two independent objects from the same class:
student1 = Student("Anna", 2)
student1.add_grade(8)
student1.add_grade(9)

print(f"{student1.name}, year {student1.year}")
print(f"Average: {student1.average()}")
````

### Each object is independent
Two objects built from the same class do not share their data. `student1` and `student2` each keep their own grades.

````python
student1 = Student("Anna", 2)
student2 = Student("Bob", 1)

student1.add_grade(10)
student1.add_grade(8)
student2.add_grade(5)

print(f"{student1.name} average: {student1.average()}")   # 9.0
print(f"{student2.name} average: {student2.average()}")   # 5.0
````

Notice that each object carries *both* its data (`name`, `year`, `grades`) *and* the actions it can perform (`add_grade`, `average`). That bundling is the entire point of OOP.

> ⚠️ **Common pitfalls**
>
> - Forgetting `self` as the first parameter of a method causes a `TypeError` when you call it.
> - Writing `self.name = name` vs just `name = name`: only the `self.` version stores the value **on the object**. Without `self.`, it is a throwaway local variable that vanishes when the method ends.

## 4. Controlling how an object displays: `__str__`

By default, printing an object shows an unhelpful memory reference. Defining the special method `__str__` lets you decide what `print()` shows, making your objects pleasant to work with.

````python
class Point:
    def __init__(self, x, y):
        self.x = x
        self.y = y

    def __str__(self):
        return f"Point({self.x}, {self.y})"

    def distance_from_origin(self):
        return (self.x ** 2 + self.y ** 2) ** 0.5

p = Point(3, 4)
print(p)                          # uses __str__  ->  Point(3, 4)
print(p.distance_from_origin())   # 5.0
````

## 5. Inheritance

### Intuition first
Suppose you have a general `Animal` class, and you want `Dog` and `Cat` classes. Dogs and cats share almost everything an animal has, so rather than rewrite it, they **inherit** from `Animal` and only add or change what is different. Inheritance is OOP's answer to *Do not Repeat Yourself*.

### Terminology
- The class being inherited from is the **parent** (or *base*) class.
- The class that inherits is the **child** (or *derived*) class.
- A child can **override** a parent's method to give it new behaviour.

````python
class Animal:
    def __init__(self, name):
        self.name = name

    def speak(self):
        return "..."

class Dog(Animal):            # Dog inherits from Animal
    def speak(self):          # override: give Dog its own behaviour
        return "Woof!"

class Cat(Animal):
    def speak(self):
        return "Meow!"

animals = [Dog("Rex"), Cat("Mina")]
for a in animals:
    print(f"{a.name}: {a.speak()}")
````

Both `Dog` and `Cat` got their `__init__` (and the `name` attribute) **for free** from `Animal`. They only had to specify what makes them different: their `speak` method.

> 🔭 **PyTorch preview.** In the deep-learning module you will write something like:
>
> ```python
> class MyNetwork(nn.Module):
>     def __init__(self):
>         super().__init__()
>         # define the layers here
> ```
>
> You can now decode every part of this: we define a **class** that **inherits** from `nn.Module` (the parent), give it its own `__init__`, and call `super().__init__()` to run the parent's setup first. Nothing here will be unfamiliar.

        self.balance += amount

        else:
            self.balance -= amount
