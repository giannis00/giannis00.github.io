## 1. The anatomy of a Matplotlib figure

Matplotlib is the foundational plotting library in Python. A plot is built from a **figure** (the whole canvas) containing one or more **axes** (an individual plot with its x- and y-axis). The typical workflow is: create data, call a plotting function, label everything, then show the figure.

````python
x = [1, 2, 3, 4, 5]
y = [1, 4, 9, 16, 25]

plt.figure(figsize=(6, 4))   # create a figure, set its size
plt.plot(x, y)               # draw a line
plt.title("A Simple Line Plot")
plt.xlabel("x")
plt.ylabel("y = x squared")
plt.grid(True)
plt.show()
````

## 2. Line plots: showing trends over a sequence

### When to use
Line plots are ideal when the x-axis has a natural order (time, or a continuous variable) and you want to show how a quantity changes along it. We will use them heavily later to plot a model's **loss over training epochs**.

````python
x = np.linspace(0, 2 * np.pi, 100)   # 100 points from 0 to 2 pi

plt.figure(figsize=(7, 4))
plt.plot(x, np.sin(x), label="sin(x)")
plt.plot(x, np.cos(x), label="cos(x)", linestyle="--")
plt.title("Two Curves on One Axis")
plt.xlabel("x (radians)")
plt.ylabel("value")
plt.legend()       # show the labels defined above
plt.show()
````

### Styling a line
You can control colour, line style, width, and markers to make a plot clearer or to distinguish series.

````python
x = np.arange(1, 8)
y = [3, 5, 4, 7, 6, 8, 9]

plt.figure(figsize=(7, 4))
plt.plot(x, y, color="purple", linewidth=2, marker="o", linestyle="-")
plt.title("A Styled Line with Markers")
plt.xlabel("day"); plt.ylabel("value")
plt.grid(True, alpha=0.3)
plt.show()
````

## 3. Scatter plots: showing relationships

### When to use
A **scatter plot** draws one dot per data point and is the primary tool for asking *'are these two variables related?'* An upward cloud suggests positive correlation, a downward cloud negative, a shapeless blob little relationship.

We will simulate study hours versus exam score, with some realistic noise:

````python
study_hours = np.random.uniform(0, 10, 50)        # 50 students, 0-10 hours
noise = np.random.normal(0, 8, 50)                # random variation
exam_score = 50 + 4 * study_hours + noise         # underlying trend plus noise
exam_score = np.clip(exam_score, 0, 100)          # keep within 0-100

plt.figure(figsize=(7, 4))
plt.scatter(study_hours, exam_score, alpha=0.7)
plt.title("Study Hours vs Exam Score")
plt.xlabel("Hours studied")
plt.ylabel("Exam score")
plt.show()
````

> 🧠 The upward trend is visible immediately: more study hours tend to mean higher scores, with scatter around the trend. In Module C, fitting a line through exactly this kind of cloud is what **linear regression** does.

## 4. Histograms: showing a distribution

### When to use
A **histogram** splits a single variable's range into bins and counts how many values fall in each. It answers *'what does the distribution look like?'*: where values cluster, whether they are symmetric, whether there are outliers.

````python
heights = np.random.normal(170, 10, 1000)   # 1000 heights, mean 170, sd 10

plt.figure(figsize=(7, 4))
plt.hist(heights, bins=30, edgecolor="black", alpha=0.7)
plt.title("Distribution of Heights")
plt.xlabel("Height (cm)")
plt.ylabel("Frequency")
plt.axvline(heights.mean(), color="red", linestyle="--", label=f"mean = {heights.mean():.1f}")
plt.legend()
plt.show()
````

The familiar bell shape is the **normal (Gaussian) distribution**, one of the most important distributions in statistics and machine learning. Notice how the histogram makes its shape obvious, where the raw 1000 numbers would not.

## 5. Bar charts: comparing categories

### When to use
**Bar charts** compare a quantity across distinct categories (cities, products, classes). The bar *length* encodes the value, making differences easy to compare at a glance.

````python
cities = ["Athens", "Patras", "Thessaloniki", "Heraklion"]
students = [120, 65, 90, 40]

plt.figure(figsize=(7, 4))
plt.bar(cities, students, color="steelblue")
plt.title("Students per City")
plt.xlabel("City")
plt.ylabel("Number of students")
plt.show()
````

### A horizontal bar chart
When category names are long, horizontal bars (`barh`) are often easier to read.

````python
languages = ["Python", "JavaScript", "Java", "C++", "Go"]
popularity = [85, 70, 55, 40, 25]

plt.figure(figsize=(7, 4))
plt.barh(languages, popularity, color="teal")
plt.title("Relative Popularity")
plt.xlabel("score")
plt.show()
````

## 6. Multiple plots in one figure

Often you want several plots side by side for comparison. `plt.subplots(rows, cols)` creates a grid of axes.

````python
fig, axes = plt.subplots(1, 2, figsize=(11, 4))

axes[0].scatter(study_hours, exam_score, alpha=0.7, color="green")
axes[0].set_title("Relationship")
axes[0].set_xlabel("Hours")
axes[0].set_ylabel("Score")

axes[1].hist(exam_score, bins=15, edgecolor="black", alpha=0.7, color="orange")
axes[1].set_title("Score distribution")
axes[1].set_xlabel("Score")
axes[1].set_ylabel("Frequency")

plt.tight_layout()   # prevent labels from overlapping
plt.show()
````

> ⚠️ **Common pitfalls**
>
> - Forgetting `plt.show()` (in some environments the figure will not render).
> - Leaving plots unlabelled: always add a title and axis labels; a figure without them cannot be interpreted.
> - Choosing the wrong plot type for the question (for example a line plot for unordered categories). Match the plot to the question.

## 🔑 Key takeaways

- Visualisation reveals structure that raw numbers hide; it is a core analytical step.
- **Line** plots show trends over an ordered axis; **scatter** plots show relationships between two variables.
- **Histograms** show the distribution of one variable; **bar charts** compare categories.
