## 2. Displaying output with `print()`

### What it does
`print()` is a **function**: a named, reusable action. Whatever you place inside its parentheses, it displays on the screen. This is how a program *communicates* with us.

### The anatomy of the statement
```
print( "Hello World!" )
  |          |
  |          value to display (here, a piece of text)
  function name
```

Text values are called **strings** and must always be wrapped in **quotes**, either double `"..."` or single `'...'` (just be consistent). You can print several items at once by separating them with commas, and Python inserts a space between them.

````python
print("Python is", "powerful", "and readable")
print("Numbers work too:", 2, 3, 4)
print("You can mix text and numbers:", "the answer is", 42)
````

### Printing numbers and simple calculations
You do not even need quotes for numbers. Python can print the result of a calculation directly:

````python
print(2 + 2)
print(10 * 5)
print("Half of 50 is", 50 / 2)
````

### Controlling the output: `sep` and `end`
Two optional settings give you finer control. `sep` changes what goes *between* items (default is a space); `end` changes what goes at the *end* of the line (default is a new line).

````python
# sep puts a custom separator between the items
print("2026", "01", "15", sep="-")      # 2026-01-15
print("a", "b", "c", sep=" | ")          # a | b | c

# end controls what finishes the line; here we keep two prints on one line
print("Loading", end="... ")
print("done!")
````

### Printing blank lines
An empty `print()` with nothing inside just prints a blank line. This is handy for spacing out your output.

````python
print("First block of text.")
print()                       # a blank line
print("Second block of text.")
````

> ⚠️ **Common pitfalls**
>
> - Forgetting the quotes around text: `print(Hello)` causes a `NameError`, because Python thinks `Hello` is a variable name.
> - Using curly quotes from a word processor (the slanted ones) instead of straight quotes: this causes a `SyntaxError`.
> - Forgetting the closing parenthesis: every `(` needs a matching `)`.

## 3. Variables: giving names to values

### Intuition first
Imagine a labelled box. You put a value inside and write a name on the label. Later, whenever you mention that name, you get back whatever is in the box. A **variable** is exactly that: a name that stores a value so you can reuse it.

### Formal definition
A variable is created with the **assignment operator** `=`. Read `age = 30` as *'let the name `age` refer to the value `30`'*. It is **not** mathematical equality. It is an instruction: *store this value under this name*.

### Why it matters
Variables let programs remember things and let us write code once and reuse values many times. Every program you will ever write, including the neural networks later in this course, is built on storing and updating values in variables.

````python
name = "Ioannis"
age = 30
height = 1.88

print("Name:", name)
print("Age:", age)
print("Height:", height)
````

### A variable can change
The whole point of a *variable* is that its value can vary. A later assignment simply replaces the old value. The variable always holds whatever was put in it most recently.

````python
score = 10
print("Start:", score)

score = 25          # the old value 10 is replaced
print("After update:", score)

score = score + 5   # take the current value, add 5, store it back
print("After adding 5:", score)
````

> 🧠 The line `score = score + 5` confuses beginners at first, because it is not an equation. Read it as a recipe: *compute the right-hand side first (the current score plus 5), then store that result back into score*. Python even has a shorthand for this: `score += 5` means exactly the same thing.

````python
count = 0
count += 1     # same as: count = count + 1
count += 1
print("Count is now", count)   # 2

price = 100
price -= 20    # same as: price = price - 20
price *= 2     # same as: price = price * 2
print("Price is now", price)   # 160
````

### Using variables together
The real power appears when you combine several variables in a calculation. Change one value and re-run, and everything that depends on it updates.

````python
price_per_item = 4
quantity = 7
total = price_per_item * quantity

print("You bought", quantity, "items.")
print("Total cost:", total)
````

### Swapping two variables
A small but elegant example: Python lets you swap two values in a single line, with no temporary variable.

````python
a = "left"
b = "right"
print("Before:", a, b)

a, b = b, a       # swap them in one line
print("After: ", a, b)
````

### Naming rules and conventions
- A name must start with a letter or an underscore `_`, never a digit.
- No spaces are allowed; use underscores, for example `average_score`.
- Python is **case-sensitive**: `score`, `Score`, and `SCORE` are three different names.

**Good practice:** choose descriptive names. `average_grade` is far better than `m`. Code is read far more often than it is written, so clarity matters.

> ⚠️ **Common pitfalls**
>
> - Starting a name with a digit (`2nd_place = ...`) is a `SyntaxError`. Put the digit later: `place_2`.
> - Using a space in a name (`my score = 5`) fails. Use `my_score`.
> - Mixing up similar names with different capitalisation, such as `Total` and `total`, leads to confusing bugs.

## 4. Core data types

Every value in Python has a **type**, which determines what you can do with it. Adding two numbers and joining two pieces of text mean different things, and the type is how Python knows the difference. The four foundational types are:

| Type | What it represents | Examples |
|------|--------------------|----------|
| `int` | whole number (integer) | `42`, `-7`, `0` |
| `float` | number with a decimal point | `3.14`, `1.82`, `-0.5` |
| `str` | text (a *string* of characters) | `"Python"`, `'AI'` |
| `bool` | a logical truth value | `True`, `False` |

The built-in `type()` function tells you the type of any value:

````python
print(type(42))
print(type(3.14))
print(type("some text"))
print(type(True))
````

You can also check the type of a value stored in a variable. This is useful when a program is not behaving as expected.

````python
x = 5
y = 5.0
z = "5"

print("x is", type(x))   # int
print("y is", type(y))   # float (because of the decimal point)
print("z is", type(z))   # str  (because of the quotes)
````

### Converting between types
Sometimes you need to change a value from one type to another. This is called **type casting** and it will be essential when reading user input, which always arrives as text.

````python
text_number = "25"              # this is a STRING, not a number
real_number = int(text_number)  # convert it to an integer

print(text_number + text_number)  # string + string joins: "2525"
print(real_number + real_number)  # int + int adds: 50
````

Here are the three casting functions you will use most. Notice what each one does and where it can fail.

````python
print(int(3.9))        # float to int: 3 (it truncates, it does NOT round)
print(float(7))        # int to float: 7.0
print(str(42))         # int to string: "42", now you can join it to text

age = 30
message = "I am " + str(age) + " years old"   # str() lets us join a number to text
print(message)
````

> ⚠️ **Common pitfalls**
>
> - `"5" + 3` raises a `TypeError`: you cannot add a string and a number. Convert first with `int(...)` or `float(...)`.
> - `int("3.5")` fails, because `int()` cannot parse a decimal string directly. Use `float("3.5")` instead.
> - `int(3.9)` gives `3`, not `4`. It removes the decimal part rather than rounding. Use `round(3.9)` if you want `4`.

## 5. Arithmetic operations

Python works as a sophisticated calculator. Note especially the difference between `/`, `//`, and `%`, which often surprises beginners.

````python
a = 10
b = 3

print("Addition:           ", a + b)
print("Subtraction:        ", a - b)
print("Multiplication:     ", a * b)
print("Division:           ", a / b)    # always a float: 3.333...
print("Floor division:     ", a // b)   # discards the remainder: 3
print("Modulo (remainder): ", a % b)    # what is left over: 1
print("Exponent (power):   ", a ** b)   # 10 to the power of 3: 1000
````

### Operator precedence
Python follows the usual mathematical order of operations: powers first, then multiplication and division, then addition and subtraction. Use **parentheses** to make your intention clear and to override the default order.

````python
print(2 + 3 * 4)      # multiplication first: 2 + 12 = 14
print((2 + 3) * 4)    # parentheses first: 5 * 4 = 20
print(10 - 2 ** 3)    # power first: 10 - 8 = 2
print(100 / 10 / 2)   # left to right: 10 / 2 = 5.0
````

### A worked example: splitting a bill
Let us combine variables and arithmetic in a realistic mini calculation. Three friends share a restaurant bill and add a tip.

````python
bill = 84.0
people = 3
tip_percent = 10

tip = bill * tip_percent / 100
total = bill + tip
per_person = total / people

print("Tip:", tip)
print("Total with tip:", total)
print("Each person pays:", per_person)
````

### The modulo operator is more useful than it looks
`%` gives the remainder of a division. Two very common uses: testing whether a number is even, and extracting the last digit.

````python
number = 17

print("Is it even?", number % 2 == 0)   # remainder 0 means even -> False here
print("Last digit:", number % 10)        # the remainder when divided by 10 -> 7

print(8 % 2)    # 0  -> 8 is even
print(9 % 2)    # 1  -> 9 is odd
````

## 6. Working with text (strings)

Strings can be **joined** with `+` (called *concatenation*) and **repeated** with `*`.

````python
first = "Ioannis"
last = "Tsioulis"

full = first + " " + last   # note the explicit space in the middle
print(full)
print("=" * 30)             # repeat the "=" character 30 times
print("ha" * 3)             # haha ha -> "hahaha"
````

### The length of a string, and reading one character
`len()` tells you how many characters a string contains. You can also read a single character by its position, counting from **0** (much more on this in Notebook 03).

````python
word = "Python"

print("Length:", len(word))   # 6 characters
print("First letter:", word[0])   # P  (positions start at 0)
print("Third letter:", word[2])   # t
print("Last letter:", word[-1])   # n  (negative counts from the end)
````

### f-strings: the modern, readable way

Placing the letter `f` immediately before the opening quote creates an **f-string**. Inside it, anything in curly braces `{ }` is evaluated and inserted into the text. This is the dominant style in modern Python and we will use it throughout the course.

````python
name = "Ioannis"
notebooks = 18

print(f"{name} prepared {notebooks} notebooks for this course.")
print(f"Seven squared is {7 ** 2}.")          # you can compute inside the braces
````

### Formatting numbers inside f-strings
A colon inside the braces lets you control how a number is displayed: the number of decimals, a percentage, or a fixed width. This produces clean, professional output.

````python
pi = 3.14159265
ratio = 0.8472

print(f"Pi to two decimals: {pi:.2f}")       # 3.14
print(f"Pi to four decimals: {pi:.4f}")      # 3.1416
print(f"As a percentage: {ratio:.1%}")        # 84.7%
print(f"Padded to width 8: |{42:8}|")         # right-aligned in 8 spaces
````

## 7. Reading input from the user

The `input()` function pauses the program and waits for the user to type something. **Important:** it *always* returns a **string**, even if the user types digits. To do maths with it, you must cast it with `int()` or `float()` first.

> Run the cell below and type your answer in the box that appears.

````python
name = input("What is your name? ")
print(f"Hello, {name}! Welcome to Python.")
````

Now an example that reads a **number**. Notice the `int(...)` wrapped around `input(...)`: without it, the value would be text and the arithmetic would fail.

````python
birth_year = int(input("Which year were you born? "))
age = 2026 - birth_year
print(f"In 2026 you turn about {age}.")
````

> ⚠️ **Common pitfalls**
>
> - Forgetting to cast input: `age = input(...)` then `age + 1` fails, because `age` is still text.
> - In a browser-based notebook, `input()` shows a small prompt box. If nothing seems to happen, look for it.

## 8. Putting it all together

Let us combine everything from this notebook into one small, complete program. It reads two numbers from the user and reports several facts about them. Read each line and connect it to the section it came from.

````python
# A small program using variables, casting, arithmetic, and f-strings
a = int(input("Enter the first whole number: "))
b = int(input("Enter the second whole number: "))

print(f"Sum:        {a + b}")
print(f"Difference: {a - b}")
print(f"Product:    {a * b}")
print(f"a divided by b: {a / b:.2f}")
print(f"Is a even? {a % 2 == 0}")
print(f"The larger value is {max(a, b)} and the smaller is {min(a, b)}")
````

> 🧠 `max()` and `min()` are two more handy built-in functions: they return the largest and smallest of the values you give them. Python comes with many such built-ins, and you will meet more as the course goes on.
