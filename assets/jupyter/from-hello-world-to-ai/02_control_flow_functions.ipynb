## 1. Making decisions: `if` / `elif` / `else`

### Intuition first
Programs constantly face forks in the road: *if it is raining, take an umbrella; otherwise, do not.* A **conditional statement** lets your code choose between different actions depending on the situation.

### Formal definition
An `if` statement runs a block of code **only when** a condition is `True`. A condition is any expression that evaluates to a boolean (`True` or `False`). You can chain alternatives with `elif` ('else if') and provide a fallback with `else`.

### Indentation defines the block
In Python, the lines that 'belong to' an `if` are marked by **indentation** (we use **4 spaces**). This is not decoration: it is how Python knows where the block starts and ends. The colon `:` at the end of the `if` line is required.

````python
temperature = 28

if temperature > 30:
    print("It is hot outside.")
elif temperature > 20:
    print("The weather is pleasant.")
else:
    print("It is cold.")
````

Trace the logic: Python checks each condition **from top to bottom** and runs the block of the **first** one that is `True`, then skips the rest. With `temperature = 28`, the first test (`> 30`) is `False`, the second (`> 20`) is `True`, so only the middle line runs. Change the value of `temperature` above and re-run to see the other branches fire.

### A simpler example: just `if`
You do not always need `elif` and `else`. Sometimes you only want to act when one condition holds.

````python
balance = -50

if balance < 0:
    print("Warning: your account is overdrawn!")

print("This line always runs, because it is not indented under the if.")
````

### Comparison and logical operators

| Comparison | Meaning | | Logical | Meaning |
|------------|---------|--|---------|---------|
| `==` | equal to | | `and` | both must be true |
| `!=` | not equal to | | `or` | at least one true |
| `>` `<` | greater / less than | | `not` | reverses true/false |
| `>=` `<=` | greater/less or equal | | | |

> ⚠️ Remember the distinction from Notebook 01: `=` **assigns** a value, while `==` **tests** for equality. Mixing them up is one of the most common beginner bugs.

Let us see each comparison produce a boolean directly:

````python
print(5 == 5)     # True
print(5 != 3)     # True
print(10 > 20)    # False
print(7 <= 7)     # True
print("cat" == "Cat")   # False (capital C is different)
````

And the logical operators combine conditions:

````python
age = 22
has_ticket = True

print(age >= 18 and has_ticket)   # both true -> True
print(age < 12 or has_ticket)     # at least one true -> True
print(not has_ticket)             # reverses True -> False
````

### A worked example: a grading scale
Conditions shine when sorting a value into categories. Here we turn a numeric grade into a letter.

````python
grade = 84

if grade >= 90:
    letter = "A"
elif grade >= 80:
    letter = "B"
elif grade >= 70:
    letter = "C"
elif grade >= 50:
    letter = "D"
else:
    letter = "F"

print(f"A grade of {grade} earns letter {letter}.")
````

### Combining conditions in a decision
Here both requirements must hold for the student to pass.

````python
grade = 7
attendance = 12

if grade >= 5 and attendance >= 10:
    print("Passed the course.")
else:
    print("Requirements not met.")
````

> ⚠️ **Common pitfalls**
>
> - Writing `if x = 5:` instead of `if x == 5:` causes a `SyntaxError`.
> - Forgetting the colon `:` at the end of the `if`/`elif`/`else` line.
> - Inconsistent indentation (mixing tabs and spaces) causes an `IndentationError`. Pick spaces and stay consistent.

## 2. Repetition with `for` loops

### Intuition first
Suppose you must greet thirty students. You would not write thirty `print` statements. You would say *'for each student, print a greeting.'* That is precisely what a **`for` loop** does: it repeats a block of code once for every item in a sequence.

### The `range()` helper
`range(n)` generates the numbers `0, 1, 2, ..., n-1`. It is the standard way to repeat something a fixed number of times.

````python
# range(5) produces 0, 1, 2, 3, 4
for i in range(5):
    print(f"Iteration number {i}")
````

### `range()` with a start, stop, and step
`range` is flexible. You can give it a starting point, a stopping point (excluded), and a step size.

````python
for i in range(1, 6):       # 1 to 5 (6 is excluded)
    print(i, end=" ")
print()                      # blank line

for i in range(0, 11, 2):   # 0 to 10 in steps of 2
    print(i, end=" ")
print()

for i in range(10, 0, -1):  # count DOWN from 10 to 1
    print(i, end=" ")
print()
````

````python
# A for loop can iterate directly over a list of items
students = ["Anna", "George", "Maria"]

for student in students:
    print(f"Welcome, {student}!")
````

### Iterating over the characters of a string
A string is also a sequence, so you can loop over its letters one by one.

````python
for letter in "PYTHON":
    print(letter)
````

### A worked example: a running total
A very common pattern: start an accumulator at 0, then add to it on each pass through the loop.

````python
numbers = [10, 25, 3, 42, 7]
total = 0

for n in numbers:
    total = total + n      # add each number to the running total

print("The sum is", total)
print("The average is", total / len(numbers))
````

> 🧠 The loop variable (here `i`, `student`, `letter`, or `n`) takes the next value from the sequence on each pass. You choose its name, so pick something meaningful.

## 3. Repetition with `while` loops

### When to use it
A `for` loop is ideal when you know *how many times* to repeat. A **`while` loop** is for when you repeat **as long as a condition holds**, possibly an unknown number of times (for example, *keep asking until the user types a valid number*).

### The danger of infinite loops
Something inside the loop must eventually make the condition `False`, otherwise the loop never stops. Below, we decrease `countdown` on every pass so it is guaranteed to reach `0`.

````python
countdown = 3

while countdown > 0:
    print(f"Countdown: {countdown}")
    countdown = countdown - 1   # shorthand: countdown -= 1

print("Lift-off!")
````

### A worked example: doubling until a limit
Here we do not know in advance how many steps are needed, which is exactly when `while` is the right tool.

````python
value = 1
steps = 0

while value < 100:
    value = value * 2
    steps = steps + 1
    print(f"After step {steps}: value = {value}")

print(f"It took {steps} doublings to pass 100.")
````

> ⚠️ **Common pitfalls**
>
> - Forgetting to update the condition variable inside a `while` loop creates an **infinite loop**. If that happens in Jupyter, interrupt the kernel with the stop button.

### Controlling loops: `break` and `continue`
- `break` immediately **exits** the loop entirely.
- `continue` **skips the rest of the current pass** and jumps to the next iteration.

````python
for n in range(1, 11):
    if n == 7:
        print("Found 7, stopping.")
        break
    if n % 2 == 0:
        continue          # skip even numbers
    print(f"Odd number: {n}")
````

## 4. Functions

### Intuition first
A **function** is a named, reusable block of code, like a recipe you can follow whenever you need it. Instead of rewriting the same steps in many places, you define them once and *call* the function by name. You have already been using built-in functions: `print()`, `input()`, `type()`, `range()`, `len()`, `max()`.

### Why it matters
Functions are the single most important tool for keeping code organised, readable, and free of repetition. Every serious program, and every machine-learning project in this course, is built from functions. The principle is **DRY**: *Do not Repeat Yourself*.

### Anatomy of a function
We define a function with the `def` keyword. The text in triple quotes is a **docstring**: a short description of what the function does.

````python
def greet(name):
    """Print a friendly greeting for the given name."""
    print(f"Good evening, {name}!")

# "Calling" the function runs its code, with a specific value for name:
greet("Ioannis")
greet("Anna")
````

Notice how we defined the greeting logic **once** but used it **twice**, with different inputs. That reuse is the whole point.

### Returning values with `return`

Printing and **returning** are different. `print()` shows something on screen; `return` hands a value **back to the caller** so it can be stored and used further. A function that computes a result should almost always `return` it.

````python
def circle_area(radius):
    pi = 3.14159
    return pi * radius ** 2

area = circle_area(5)          # the returned value is stored in "area"
print(f"The area is {area:.2f}")

print(f"A radius-10 circle has area {circle_area(10):.2f}")  # use it directly too
````

> 🧠 The difference between `print` and `return` is subtle but important. A function that *prints* shows you a value but hands nothing back, so `result = printing_function()` would leave `result` empty. A function that *returns* gives a value you can store, reuse, and combine. Try the next cell to feel the difference.

````python
def add_and_print(a, b):
    print(a + b)        # shows the sum but returns nothing

def add_and_return(a, b):
    return a + b        # hands the sum back

x = add_and_print(2, 3)      # prints 5
y = add_and_return(2, 3)     # prints nothing, but stores 5 in y

print("x holds:", x)         # None (nothing was returned)
print("y holds:", y)         # 5
print("We can keep using y:", y * 10)
````

### Multiple parameters and returning several values
A function can take several inputs and even hand back several results at once (as a tuple, which you will meet properly in Notebook 03).

````python
def stats(a, b, c):
    total = a + b + c
    average = total / 3
    return total, average      # return two values together

the_sum, the_avg = stats(10, 20, 30)
print("Sum:", the_sum)
print("Average:", the_avg)
````

### Default parameter values
A parameter can have a **default**, used when the caller does not supply one.

````python
def greet(name, language="English"):
    if language == "English":
        print(f"Hello, {name}!")
    else:
        print(f"Bonjour, {name}!")

greet("Maria")                      # uses the default language
greet("Marie", language="French")   # overrides the default
````

### A worked example: a reusable helper
Once a function exists, you can call it many times with different inputs. Here is a small temperature converter.

````python
def celsius_to_fahrenheit(c):
    return c * 9/5 + 32

for temp in [0, 20, 37, 100]:
    print(f"{temp} C = {celsius_to_fahrenheit(temp)} F")
````

## 5. Variable scope

### Intuition first
A variable created **inside** a function lives only inside that function, like notes on a whiteboard that get wiped when the meeting ends. This is called **local scope**, and it is a *feature*: it keeps functions self-contained and prevents them from accidentally interfering with each other.

````python
def example():
    local_message = "I exist only inside this function"
    print(local_message)

example()

# The line below would raise a NameError, because local_message
# does not exist outside the function. Uncomment it to see:
# print(local_message)
````

## 6. Handling errors with `try` / `except`

### Why it matters
Real programs face messy input: a user types `"hello"` when you expected a number, a file is missing, a network call fails. Without protection, the program **crashes**. The `try` / `except` mechanism lets you *catch* an error and respond calmly instead.

Put the risky code in the `try` block; if an error of the named type occurs, the `except` block runs instead of crashing.

````python
try:
    number = int(input("Enter a number: "))
    print(f"Its square is {number ** 2}")
except ValueError:
    print("That was not a valid number.")
````

> 🧠 `ValueError` is the specific error Python raises when `int()` cannot interpret the text as a number. Catching specific error types (rather than all errors blindly) is good practice: it makes bugs easier to find.

## 🔑 Key takeaways

- `if` / `elif` / `else` let code make decisions; **indentation** (4 spaces) defines each block.
