## 1. Lists

### Intuition first
A **list** is an ordered shopping list of items. You can read any item by its position, add new ones, remove ones, and reorder them. It is the most frequently used container in Python.

### Formal definition
A list is an **ordered, mutable** collection written inside square brackets `[ ]`. *Ordered* means each element has a fixed position; *mutable* means the list can be changed after creation. A list can hold any types, even mixed.

### Indexing: reading by position
Positions (called **indices**) start at **0**, not 1. This trips up almost every beginner, so internalise it now: the first element is at index `0`, the second at `1`, and so on. Negative indices count from the end.

````python
grades = [8, 6, 9, 7, 10]

print("The whole list:", grades)
print("First element (index 0):", grades[0])
print("Second element (index 1):", grades[1])
print("Last element (index -1):", grades[-1])
print("Second to last (index -2):", grades[-2])
print("Number of elements:", len(grades))
````

### Lists can hold mixed types, and other lists
A list is not limited to one type. It can even contain other lists, which is how we represent tables and grids.

````python
mixed = ["Anna", 21, 8.5, True]
print(mixed)
print("The name is", mixed[0], "and the age is", mixed[1])

# A list of lists (think of it as rows of a table)
grid = [[1, 2, 3],
        [4, 5, 6]]
print("Row 0:", grid[0])
print("The value in row 1, column 2:", grid[1][2])   # 6
````

### Slicing: reading a sub-section
The syntax `my_list[start:stop]` returns a new list from index `start` up to **but not including** `stop`. This 'stop is exclusive' rule is consistent with how `range()` works.

````python
letters = ["a", "b", "c", "d", "e"]

print(letters[1:3])   # indices 1 and 2  -> [b, c]   (3 is excluded)
print(letters[:2])    # from the start   -> [a, b]
print(letters[2:])    # to the end       -> [c, d, e]
print(letters[-2:])   # last two         -> [d, e]
````

> ⚠️ **Common pitfalls**
>
> - Off-by-one errors: `grades[5]` on a 5-element list raises `IndexError`, because valid indices are 0 to 4.
> - Remembering that the `stop` index in a slice is **excluded**: `letters[1:3]` gives two items, not three.

### Modifying lists
Lists offer many methods. Here are the most common, shown together:

````python
fruits = ["apple", "banana"]
print("Start:", fruits)

fruits.append("orange")        # add to the end
print("After append:", fruits)

fruits.insert(0, "strawberry") # insert at a specific index
print("After insert:", fruits)

fruits.remove("banana")        # remove by value
print("After remove:", fruits)

last = fruits.pop()            # remove AND return the last item
print("Popped:", last, "| remaining:", fruits)
````

### Looping over a list, with and without the index
Often you want both the position and the value. The built-in `enumerate()` gives you both at once.

````python
colours = ["red", "green", "blue"]

# Just the values
for colour in colours:
    print(colour)

print("---")

# Position and value together with enumerate()
for index, colour in enumerate(colours):
    print(f"{index}: {colour}")
````

### Sorting and other useful operations

````python
numbers = [4, 1, 8, 3, 9, 2]

print("Sorted copy:", sorted(numbers))     # returns a new sorted list
print("Original unchanged:", numbers)
print("Largest:", max(numbers))
print("Smallest:", min(numbers))
print("Sum:", sum(numbers))
print("Is 8 in the list?", 8 in numbers)   # membership test -> True
````

## 2. Tuples

### Intuition first
A **tuple** is like a list that has been 'sealed': once created, it cannot be changed. Use it for collections that should never be modified, such as geographic coordinates, RGB colours, or a date as `(year, month, day)`.

### Why immutability is useful
Because tuples cannot change, Python can treat them as safe, fixed records. They also make your intent clear to anyone reading the code: *this group of values belongs together and must not be altered.* Tuples use parentheses `( )`.

````python
point = (3, 5)
x, y = point          # "unpacking": assign each element to a variable
print(f"x = {x}, y = {y}")

colour = (255, 100, 0)   # an RGB colour
print("Red component:", colour[0])

# point[0] = 10  ->  would raise a TypeError: tuples cannot be modified
````

## 3. Dictionaries

### Intuition first
A real dictionary maps a **word** to its **definition**. A Python **dictionary** maps a **key** to a **value**. Instead of looking things up by numeric position (as in a list), you look them up by a meaningful key.

### Why it matters
Dictionaries are arguably the most important practical data structure. They appear everywhere: dataset records, model configuration, data from web services, counting occurrences. Mastering them pays off immediately.

### Formal definition
A dictionary stores **key to value** pairs inside curly braces `{ }`. Keys must be unique and immutable (strings and numbers are typical). Values can be anything, including lists or other dictionaries.

````python
student = {
    "name": "Anna",
    "year": 2,
    "grades": [8, 9, 7],
}

print(student["name"])             # look up a value by its key
print(student["grades"])           # values can be lists
student["email"] = "anna@uni.edu"  # add a new key-value pair
print(student)
````

### Safe lookups with `.get()`
Asking for a key that does not exist with `student["phone"]` crashes the program. The `.get()` method returns `None` (or a default you choose) instead, which is much safer.

````python
print(student.get("name"))            # Anna
print(student.get("phone"))           # None (the key does not exist)
print(student.get("phone", "N/A"))    # supply a default instead of None
````

````python
# Iterating over a dictionary with .items() gives you key and value together
for key, value in student.items():
    print(f"{key}: {value}")

print("---")
print("Just the keys:", list(student.keys()))
print("Just the values:", list(student.values()))
````

### A worked example: a small phone book
Dictionaries are perfect for lookups by name. Here we build one and query it.

````python
phonebook = {
    "Alice": "555-1234",
    "Bob": "555-5678",
    "Carol": "555-9999",
}

name = "Bob"
if name in phonebook:
    print(f"{name}'s number is {phonebook[name]}")
else:
    print(f"{name} is not in the phone book.")
````

> ⚠️ **Common pitfalls**
>
> - Accessing a missing key with `student["phone"]` raises a `KeyError`. Use `student.get("phone")` to get `None` instead of an error.
> - Keys are unique: assigning to an existing key **overwrites** its value rather than adding a second entry.

## 4. Sets

### Intuition first
A **set** is an unordered bag of **unique** items: duplicates are automatically discarded. It mirrors the mathematical notion of a set and supports the same operations: union, intersection, difference.

### A very common use
The fastest way to remove duplicates from a list is to convert it to a set.

````python
numbers = [1, 2, 2, 3, 3, 3, 4]
unique = set(numbers)
print("Unique values:", unique)
print("How many unique:", len(unique))
````

````python
a = {1, 2, 3}
b = {3, 4, 5}
print("Intersection (common):", a & b)
print("Union (all):", a | b)
print("Difference (in a only):", a - b)

# A practical example: which students are in both clubs?
chess = {"Anna", "Bob", "Carol"}
drama = {"Bob", "Carol", "Dan"}
print("In both clubs:", chess & drama)
````

## 5. List comprehensions

### Intuition first
Very often we build a new list by transforming or filtering an existing sequence. A **list comprehension** expresses that idea in a single, readable line. It is one of Python's most beloved features and you will see it constantly in data-science code.

### How to read it
`[expression for item in sequence]` reads as *'give me `expression` for each `item` in `sequence`.'* You can append `if condition` to keep only some items.

````python
# The long way:
squares = []
for n in range(1, 6):
    squares.append(n ** 2)
print(squares)

# The comprehension way: identical result, one line
squares = [n ** 2 for n in range(1, 6)]
print(squares)
````

Here are a few more, each showing a different shape of the same idea:

````python
# With a filter: keep only even numbers
evens = [n for n in range(10) if n % 2 == 0]
print(evens)

# Transform text: uppercase every word
words = ["python", "is", "fun"]
shouted = [w.upper() for w in words]
print(shouted)

# Combine transform and filter: lengths of the long words only
names = ["Al", "Beatrice", "Cy", "Deborah"]
long_lengths = [len(n) for n in names if len(n) > 3]
print(long_lengths)
````

> 🧠 Introduce a comprehension to yourself as 'the same loop, compressed'. If a comprehension ever becomes hard to read, it is perfectly fine to write it as an ordinary loop instead. Readability beats cleverness.

## 6. Essential string methods

Strings come with built-in **methods** for cleaning and transforming text, skills you will need constantly when preparing real-world data, which is almost always messy.

````python
text = "  Machine Learning with Python  "

print("Original:", repr(text))
print("Stripped:", repr(text.strip()))       # remove leading/trailing spaces
print("Lower:", text.lower())                 # to lowercase
print("Upper:", text.upper())                 # to uppercase
print("Replaced:", text.replace("Python", "PyTorch"))
````

````python
# split() breaks a string into a list; join() glues a list into a string
sentence = "one,two,three"
parts = sentence.split(",")
print("Split:", parts)

date = "-".join(["2026", "01", "15"])
print("Joined:", date)

# Some useful tests
print("HELLO".startswith("HE"))   # True
print("report.pdf".endswith(".pdf"))   # True
print("Python3".isalpha())        # False (contains a digit)
````

> 🧠 `split()` and `join()` are inverses of each other and appear constantly in data cleaning: `split` breaks text apart, `join` glues pieces back together.
